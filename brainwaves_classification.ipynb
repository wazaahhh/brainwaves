### Methodology

#### Input

The power spectrum every (half?) second as calculated by the Neurosky software with the output of the Neurosky Mindset headset. Power spectra, for which a sufficient data acquisition quality (c.f. Benjamin code) was not met, were discarded

#### Output

(log) binned probability distribution function of brain frequency activations, aggregated over the number of power spectra generated by the Neurosky Software. 

#### Process

Takes the median value over all available power spectra in one sample, for each quarter frequency range over all sampled frequencies (4 values per hertz over 256 Hz, i.e., 1024 values). This gives a probability density function (pdf) of finding a frequency activation for a given sample.

The pdf is basically the input for the classifier, but it makes sense to reduce the vector length by binning the data. The binning is performed as follow. We take the value the logarithmic values of both axes [frequency, pdf(frequency)], and we perform a standard binning, by taking the median values of all points in each bin (see Figure above).
