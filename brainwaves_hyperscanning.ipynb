### Power Spectrum and Generalized Entropy

```python
def pSpectrum(vector):
    
    A = np.fft.fft(vector)
    ps = np.abs(A)**2
    ps = ps[:len(ps)/2]
        
    return ps

#ps = pSpectrum(test[500:600])
#loglog(ps)

def entropy(power_spectrum,q):
    q = float(q)
    
    power_spectrum = np.array(power_spectrum)
        
    if not q ==1:
        S = 1/(q-1)*(1-np.sum(power_spectrum**q))
    else:
        S = - np.sum(power_spectrum*np.log2(power_spectrum))
        
    return S

def entropySeries(time,rawData,q,windowSize=1,normalize=True):
    
    #t = tSeries['rawTime']
    #raw = tSeries['rawValue']
    
    t = time
    raw = rawData
    
    intervales = np.arange(np.trunc(min(t)),np.trunc(max(t)+1),windowSize)

    T = []
    S = []
    
    for i,ix in enumerate(intervales[:-1]):
        #print i,ix
        
        c = (t>=ix)*(t<intervales[i+1])
        
        vector = raw[c]
        ps = np.array(pSpectrum(vector)) # Power Spectrum
        
        #if normalize:
        ps = ps/np.sum(ps)
        
        s = entropy(ps,q) #Compute Entropy
        S.append(s)
        T.append(i)

    S = np.array(S)
    T = np.array(T)
    
    if normalize:
        S = (S-np.mean(S))/np.std(S)
        
    return {'time' : T, 'S': S}
```

### Process CSV

```python
def processRawSignal(csvType):
    data = {'timestamp' : [],'poorSignal' : [], 'meditation' : {'mTime' : [], 'mValue' : []}, 'attention' : {'aTime' : [], 'aValue' : []}, 'raw' : {'rawTime' : [], 'rawValue' : []}}
    types = ['PoorSignal','Meditation','Attention','Raw'] 

    for l, line in enumerate(csvType):
        try:
            typ = line[-2][:-1]
        except:
            continue
            
        if typ=='Meditation':
            data['meditation']['mTime'].append(line[0][:-1])
            data['meditation']['mValue'].append(line[-1])
        
        if typ=="Attention":
            data['attention']['aTime'].append(line[0][:-1])
            data['attention']['aValue'].append(line[-1])
            
        if typ=="Raw":
            data['raw']['rawTime'].append(line[0][:-1])
            data['raw']['rawValue'].append(int(line[-1]))
    return data


def makeArrays(data):
    
    data['meditation']['mTime'] = np.array(map(float,data['meditation']['mTime']))
    data['meditation']['mValue'] = np.array(map(float,data['meditation']['mValue']))
        
    data['attention']['aTime'] = np.array(map(float,data['attention']['aTime']))
    data['attention']['aValue'] = np.array(map(float,data['attention']['aValue']))
            
    data['raw']['rawTime'] = np.array(map(float,data['raw']['rawTime']))
    data['raw']['rawValue'] = np.array(map(float,data['raw']['rawValue']))
    
    return data
    

def loadFromS3(date,session,subject,loadJson=True):
    
    #print "destination : ", "%s/%s/%s/%s.json"%(date,session,name,subject)
    
    if loadJson:
        try:
            key = bucket.get_key("%s/%s/%s.json"%(date,session,subject))
            data = key.get_contents_as_string()
            data = json.loads(data)
            data = makeArrays(data)
            
            return data
        except:
            pass
    
    key = bucket.get_key(dStructure[date][session][subject])
    d = key.get_contents_as_string()
    f = StringIO.StringIO(key.get_contents_as_string())
    csvType = csv.reader(f,delimiter=' ')
    
    data = processRawSignal(csvType)        
    jsonData = json.dumps(data)
    
    
    key = bucket.new_key("%s/%s/%s.json"%(date,session,subject))
    key.set_contents_from_string(jsonData)
    
    data = makeArrays(data)
    
    return data


def plotEntropy(date,session,subject,q=1,wSize=500):
    tSeries = loadFromS3(date,session,subject)['raw']['rawValue']
    Sq = entropySeries(tSeries,q,windowSize=wSize,normalize=False)

    title("%s (%s)"%(subject,name))
    plot(Sq['time'],Sq['S'],'b-')
    xlabel("time [a.u]")
    ylabel("Entropy (q = %s)"%q)

    return Sq


def syncSessionName(date,session,entropy=True):
    keys = dStructure[date][session].keys()
    
    data = {}
    
    minT = []
    maxT = []
    
    for subject in keys:
        
        data[subject] = loadFromS3(date,session,subject)['raw']    
        minT.append(min(data[subject]['rawTime']))
        maxT.append(max(data[subject]['rawTime']))
        
        #print subject,len(data[subject]['rawTime'])
        
    min_t = np.max(minT)
    max_t = np.min(maxT)
    
    #print min_t,max_t
    
    for subject  in keys:
        c = (data[subject]['rawTime'] > min_t)*(data[subject]['rawTime'] < max_t)
        data[subject]['rawTime'] = data[subject]['rawTime'][c]
        data[subject]['rawValue'] = data[subject]['rawValue'][c]

        print subject,"%.2f , %.2f , %.2f"%(min(data[subject]['rawTime']),max(data[subject]['rawTime']),len(data[subject]['rawTime']))
        
    return data

def syncEntropy(data,q,normalize=True):

    subjects = data.keys()
    
    try:
        subjects.remove('q')
    except:
        pass

    for subject in subjects:
        #print data[subject]['rawTime']#,data[subject]['rawValue']
        S = entropySeries(data[subject]['rawTime'],data[subject]['rawValue'],q,normalize=normalize)
        data[subject] = {'entropy' : S}
        print "%s, mean : %.2f, std : %.2f, skew : %.2f, entropy : %.2f"%(subject,np.mean(S['S']),np.std(S['S']),stats.skew(S['S']),entropy(S['S'],1))

    data['q'] = q
    
    return data


def syncPlot(dataS):
    
    subjects = dataS.keys()
    subjects.remove('q')
    
    l = len(subjects)
    
    pl.figure(1,(12,5*l))
    
    for i,subject in enumerate(subjects):
        print i,subject
        pl.subplot(100*l+10+i)
        pl.title(subject)
        t = dataS[subject]['entropy']['time']
        S = dataS[subject]['entropy']['S']
        pl.plot(t,S)
    
def syncRankCorr(dataS, range=[0,-1]):
    
    subjects = dataS.keys()
    subjects.remove('q')
    
    for i,ix in enumerate(subjects):
        for j,jx in enumerate(subjects):
            if i>j:
                corr,pvalue = stats.spearmanr(dataS[ix]['entropy']['S'][range[0]:range[1]],dataS[jx]['entropy']['S'][range[0]:range[1]])
                print "Rank Correlation between %s and %s : %.2f ( %.2f )"%(ix,jx,corr,pvalue)
    
```
